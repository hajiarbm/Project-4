```python
# The best way to deal with it is to define a function, and try to change the value to int, if the data is correct there
# will be no problem, otherwise an exception will happen which cause to substitue the value with nan
import numpy as np



def convert_column_to_int(df, column_name):
    """
    Convert a column in a DataFrame from strings to integers,
    replacing non-numeric values with NaN.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    column_name (str): The name of the column to convert.

    Returns:
    pd.DataFrame: The DataFrame with the specified column converted to integers.
    """
    def convert_value(value):
        try:
            return int(value)
        except ValueError:
            return np.nan

    df[column_name] = df[column_name].apply(convert_value)
    return df
```

```python
# Get the columns of df
column = df.columns.drop(['School Name', 'State Name [Public School] Latest available year',
       'School ID - NCES Assigned [Public School] Latest available year',
       'School Type [Public School] 2022-23',
       'Start of Year Status [Public School] 2022-23'])
```

```python
column
```

```python
# Now, for each column, we call the method to clean the data
for col in column:
    df = convert_column_to_int(df, col)
```
